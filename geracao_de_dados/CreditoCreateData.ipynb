## Indicadores e Valores

### Indicadores de Definição

##### 1. EBITDA
**Fórmula**:  
```
Lucro Operacional + Depreciação + Amortização
```

**Origem**: Balanço / DRE

**Informações**:  
Este indicador será necessário apenas para compor o cálculo da Margem de EBITDA, que será uma métrica melhor utilizada para análise de resultados financeiros.

**Parâmetros**:
- 1 (Muito ruim): EBITDA negativo ou extremamente baixo.  
- 2 (Ruim): Baixo, mas positivo, gerando algum lucro operacional, mas não o bastante para investimentos ou crescimento.  
- 3 (Regular): Moderado, cobrindo as despesas, mas com espaço para melhorias na eficiência e produtividade.  
- 4 (Bom): Alto em relação às receitas, permitindo que a empresa faça investimentos em crescimento e tenha uma posição financeira saudável.  
- 5 (Muito bom): EBITDA muito alto, com forte geração de caixa e fluxo de caixa robusto.

---

##### 2. Fluxo de Caixa Operacional - Fluxo de Caixa Livre
**Fórmula**:  
```
FCO = Fluxo de Caixa Operacional − CAPEX
```

**Detalhamento do FCO**:  
```
FCO = Lucro líquido + Ajuste não caixa + Variação no capital de giro
```
- Lucro líquido = Resultado final do DRE  
- Ajuste não caixa = Depreciação, amortização, provisões, equivalência patrimonial  
- Variação no capital de giro = Aumento de contas a receber (-), estoques (-), contas a pagar (+)

**Exemplo**:  
```
Lucro líquido = 100.000  
Depreciação = 20.000  
Aumento contas a receber = 10.000  
Aumento contas a pagar = 5.000  
FCO = 100.000 + 20.000 - 10.000 + 5.000 = 115.000
```

**CAPEX**: Despesa de capital = Ativo imobilizado (Balanço Patrimonial)

**Origem**: Balanço / DRE  
**Parâmetros**: N/A

---

##### 3. Ponto de Equilíbrio
**Fórmula**:  
```
(Custos fixos e despesas fixas) – (Depreciação, amortização e exaustão) / Margem de contribuição unitária
```

**Origem**: Balanço / DRE  
**Parâmetros**: N/A

---

##### 4. Parecer do Analista
**Informações**:
- Decisão de **não conceder crédito**
- Decisão de **conceder crédito**

Com justificativa do motivo da decisão.

**Origem**: N/A  
**Parâmetros**: Avaliação qualitativa

---

##### 5. Porte da Empresa
**Informações**:  
Faixa de faturamento, com base em classificação do BNDES.

**Origem**: BNDES

**Parâmetros**:
- MEI: Até R$ 144.900  
- 1 (Muito ruim): Microempresa - ROB ≤ R$ 350.000  
- 2 (Ruim): Pequena empresa - ROB > R$ 350.000 até R$ 3,6 milhões  
- 3 (Regular): Média empresa - ROB > R$ 3,6 milhões até R$ 90 milhões  
- 4 (Bom): Média-grande empresa - ROB > R $90 milhões até R$ 300 milhões  
- 5 (Muito bom): Grande empresa - ROB > R$ 300 milhões

---

##### 6. Possui Garantias
**Informações**:

**Garantias pessoais**:
- Aval  
- Fiança

**Garantias reais**:
- Fiança bancária  
- Alienação fiduciária  
- Penhor mercantil ou rural  
- Hipoteca

**Origem**: Análise de crédito/documentação

**Parâmetros**:
- 1 (Muito ruim): Não possui garantias  
- 5 (Muito bom): Possui garantias

> Este indicador deve ser **considerado para o limite de crédito**, **não para o rating da empresa**. A empresa pode estar financeiramente mal, mas receber crédito com base em garantias.

## Dataset

Este conjunto de dados foi criado sinteticamente para simular um sistema de análise de concessão de crédito para empresas no contexto financeiro brasileiro. O objetivo principal é avaliar a "saúde financeira" das empresas solicitantes e determinar:

* Se o crédito deve ser concedido
* Qual o limite de crédito apropriado

O modelo considera 11 indicadores chave, cada um com um peso específico na avaliação final:

* Margem EBITDA (20%): Mede quanto do faturamento se converte em lucro operacional bruto
* Dívida Líquida/EBITDA (10%): Indica em quantos anos a empresa pagaria sua dívida usando apenas o EBITDA atual
* Índice de Liquidez (10%): Avalia a capacidade de pagamento no curto prazo
* Ciclo Financeiro (10%): Mostra quantos dias a empresa opera com capital próprio investido
* Conversão de EBITDA em FCO (10%): Mede quanto do EBITDA realmente se converte em caixa operacional
* Auditoria em Demonstrações Financeiras (10%): Verifica se há auditoria independente das demonstrações contábeis
* Tempo de Atuação em Anos (5%): Considera o histórico e experiência da empresa no mercado
* Pendências Financeiras (5%): Verifica registros de dívidas não pagas
* Garantia (5%): Avalia se a empresa apresenta bens que podem ser utilizados como garantia
* Seguros (5%): Verifica se a empresa possui apólices de seguro ativas
* Serasa Score PJ (5%): Pontuação de crédito que indica probabilidade de pagamento em dia

**NOTAS**

Geração dos Indicadores Financeiros
Para cada indicador financeiro, o código segue um padrão:

* Gera valores aleatórios considerando as distribuições apropriadas
* Diferencia empresas de perfil bom e ruim usando distribuições estatísticas distintas
* Atribui uma nota (1-5) com base em faixas de valores
* Calcula a nota ponderada multiplicando a nota pelo peso do indicador

* Os indicadores são gerados de forma correlacionada (exemplo: dívida baseada no EBITDA), simulando dados reais onde as variáveis financeiras normalmente estão interrelacionadas.
* Diferenciação de Perfis: O código usa o array perfil_ruim para consistentemente gerar dados que refletem a saúde financeira global da empresa em todos os indicadores.

Esta estrutura de geração garante um conjunto de dados sintéticos que simula de forma realista o comportamento de empresas no contexto de análise de crédito, com padrões e correlações que se assemelham aos encontrados em dados reais.

````python
import pandas as pd
import numpy as np
import random
from datetime import datetime

# Função para gerar dados sintéticos de crédito empresarial no contexto brasileiro
def gerar_dados_credito_empresarial(num_amostras=100, seed=42, taxa_reprovacao_alvo=0.40):
    # Definir seed para reprodutibilidade
    np.random.seed(seed)
    random.seed(seed)

    # Lista de empresas de auditoria no Brasil
    auditorias = ['KPMG', 'Deloitte', 'Ernst & Young', 'PwC', 'Grant Thornton','Biolab', 'Não']

    # Lista de tipos de pendências financeiras
    pendencias = [
        'Sem pendências',
        'Protesto',
        'Protesto e ações judiciais',
        'Protesto e divida vencidas',
        'Dívida vencida',
        'Pefin',
        'Pefin e dívida vencida',
        'Ações judiciais'
    ]

    # Lista de tipos de seguros
    seguros = [
        'não',
        'seguro de estoque',
        'seguro patrimonial',
        'seguro de responsabilidade civil',
        'seguro empresarial completo'
    ]

    # Lista de tipos de garantia
    garantias = [
        'não',
        'imóvel',
        'veículos',
        'máquinas e equipamentos',
        'recebíveis',
        'aval dos sócios'
    ]

    # Gerar IDs de empresas
    empresas = list(range(1, num_amostras + 1))

    # Calculamos quantas amostras devem ser "ruins" para atingir a taxa de reprovação desejada
    num_ruins = int(num_amostras * taxa_reprovacao_alvo)

    # Vamos criar um array que diz quais empresas terão perfil ruim
    perfil_ruim = np.zeros(num_amostras, dtype=bool)
    perfil_ruim[:num_ruins] = True
    np.random.shuffle(perfil_ruim)

    # EBITDA - Valores em reais, distribuição log-normal para representar a realidade empresarial
    ebitda = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas com perfil ruim têm EBITDA menor
            valor = np.random.lognormal(mean=14, sigma=2, size=1)[0]
        else:
            valor = np.random.lognormal(mean=16, sigma=2, size=1)[0]
        ebitda.append(valor)

    # Dívida - Correlacionada com EBITDA, mas com variação
    divida = []
    for i, e in enumerate(ebitda):
        if perfil_ruim[i]:
            # Empresas ruins têm dívidas mais altas em relação ao EBITDA
            fator = np.random.uniform(2.0, 4.0)
        else:
            fator = np.random.uniform(0.5, 2.0)
        divida.append(e * fator)

    # Multiplicador do Setor - Constante para simplicidade
    multiplicador_setor = [7.4] * num_amostras

    # Margem EBITDA - Valores típicos entre 1% e 30%
    margem_ebitda = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins têm margem EBITDA menor
            valor = np.random.uniform(1.0, 8.0)
        else:
            valor = np.random.uniform(8.0, 30.0)
        margem_ebitda.append(valor)

    # Peso da margem EBITDA - Fixo em 20% conforme explicação
    peso_margem_ebitda = [0.2] * num_amostras

    # Nota do indicador Margem EBITDA (1-5)
    nota_i1 = []
    for margem in margem_ebitda:
        if margem < 3:
            nota = 1
        elif margem < 6:
            nota = 2
        elif margem < 9:
            nota = 3
        elif margem < 12:
            nota = 4
        else:
            nota = 5
        nota_i1.append(nota)

    # Nota ponderada = nota * peso
    nota_pi1 = [n * 0.2 for n in nota_i1]

    # Dívida líquida/EBITDA
    divida_liquida_ebitda = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins têm mais dívida em relação ao caixa
            caixa = divida[i] * np.random.uniform(0, 0.3)
        else:
            caixa = divida[i] * np.random.uniform(0.3, 1.5)

        divida_liquida = divida[i] - caixa
        if ebitda[i] > 0:  # Evitar divisão por zero
            divida_liquida_ebitda.append(divida_liquida / ebitda[i])
        else:
            divida_liquida_ebitda.append(np.random.uniform(5, 10))  # Caso de EBITDA zero ou negativo

    # Peso fixo de 10%
    peso_divida_liquida_ebitda = [0.1] * num_amostras

    # Nota do indicador Dívida líquida/EBITDA (1-5)
    nota_i2 = []
    for ratio in divida_liquida_ebitda:
        if ratio >= 2.3:
            nota = 1
        elif ratio >= 1.5:
            nota = 2
        elif ratio >= 1.0:
            nota = 3
        elif ratio >= 0.6:
            nota = 4
        else:  # Ratio negativo = excelente
            nota = 5
        nota_i2.append(nota)

    # Nota ponderada
    nota_pi2 = [n * 0.1 for n in nota_i2]

    # Índice de liquidez - Valores típicos entre 0.5 e 2.5
    indice_liquidez = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins têm índice de liquidez mais baixo
            valor = np.random.uniform(0.5, 1.0)
        else:
            valor = np.random.uniform(1.0, 2.5)
        indice_liquidez.append(valor)

    # Peso fixo de 10%
    peso_indice_liquidez = [0.1] * num_amostras

    # Nota do indicador Índice de liquidez (1-5)
    nota_i3 = []
    for il in indice_liquidez:
        if il < 0.5:
            nota = 1
        elif il < 1.0:
            nota = 2
        elif il < 1.5:
            nota = 3
        elif il <= 2.0:
            nota = 4
        else:
            nota = 5

        nota_i3.append(nota)

    # Nota ponderada
    nota_pi3 = [n * 0.1 for n in nota_i3]

    # Ciclo Financeiro (dias) - Pode ser negativo (ótimo) até muito alto (ruim)
    ciclo_financeiro = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins têm ciclo financeiro mais longo
            valor = np.random.normal(90, 40)
        else:
            valor = np.random.normal(20, 40)
        ciclo_financeiro.append(valor)

    # Peso fixo de 10%
    peso_ciclo_financeiro = [0.1] * num_amostras

    # Nota do indicador Ciclo Financeiro (1-5)
    nota_i4 = []
    for cf in ciclo_financeiro:
        if cf < -30:
            nota = 1
        elif cf < -10:
            nota = 2
        elif cf < 10:
            nota = 3
        elif cf <= 30:
            nota = 4
        else:
            nota = 5

        nota_i4.append(nota)

    # Nota ponderada
    nota_pi4 = [n * 0.1 for n in nota_i4]

    # Conversão de EBITDA em FCO (%) - Valores típicos entre -100% e 150%
    conversao_ebitda_fco = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins têm conversão de EBITDA em FCO mais baixa ou mesmo negativa
            valor = np.random.normal(30, 40)
        else:
            valor = np.random.normal(90, 30)
        conversao_ebitda_fco.append(valor)

    # Peso fixo de 10%
    peso_conversao_ebitda_fco = [0.1] * num_amostras

    # Nota do indicador Conversão de EBITDA em FCO (1-5)
    nota_i5 = []
    for conv in conversao_ebitda_fco:
        if conv < 20:
            nota = 1
        elif conv < 40:
            nota = 2
        elif conv < 60:
            nota = 3
        elif conv < 80:
            nota = 4
        else:
            nota = 5

        nota_i5.append(nota)

    # Nota ponderada
    nota_pi5 = [n * 0.1 for n in nota_i5]

    # Tempo de atuação em anos - Distribuição log-normal para ter mais empresas jovens
    tempo_atuacao = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins têm menos tempo de mercado
            valor = np.random.lognormal(mean=1.5, sigma=0.8)
        else:
            valor = np.random.lognormal(mean=2.5, sigma=0.8)
        tempo_atuacao.append(max(1, int(valor)))  # Mínimo de 1 ano

    # Peso fixo de 5%
    peso_tempo_atuacao = [0.05] * num_amostras

    # Nota do indicador Tempo de atuação (1-5)
    nota_i6 = []
    for tempo in tempo_atuacao:
        if tempo <= 3:
            nota = 1
        elif tempo <= 7:
            nota = 2
        elif tempo <= 12:
            nota = 3
        elif tempo <= 20:
            nota = 4
        else:
            nota = 5

        nota_i6.append(nota)

    # Nota ponderada
    nota_pi6 = [n * 0.05 for n in nota_i6]

    # Auditoria em Demonstrações Financeiras
    auditoria = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins tendem a não ter auditoria ou ter de menor expressão
            aud = np.random.choice(auditorias, p=[0.05, 0.05, 0.05, 0.05, 0.1, 0.1, 0.6])
        else:
            aud = np.random.choice(auditorias, p=[0.3, 0.2, 0.2, 0.2, 0.05, 0.03, 0.02])
        auditoria.append(aud)

    # Peso fixo de 10%
    peso_auditoria = [0.1] * num_amostras

    # Nota do indicador Auditoria (1-5)
    nota_i7 = []
    for aud in auditoria:
        if aud == 'Não':
            nota = 1
        elif aud == 'Biolab':
            nota = 2
        elif aud in ['Grant Thornton']:
            nota = 4
        elif aud in ['KPMG', 'Deloitte', 'Ernst & Young', 'PwC']:
            nota = 5

        nota_i7.append(nota)

    # Nota ponderada
    nota_pi7 = [n * 0.1 for n in nota_i7]

    # Pendências Financeiras
    pendencia = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins tendem a ter mais pendências
            p = np.random.choice(pendencias, p=[0.05, 0.1, 0.15, 0.15, 0.2, 0.15, 0.1, 0.1])
        else:
            p = np.random.choice(pendencias, p=[0.7, 0.1, 0.05, 0.03, 0.03, 0.03, 0.03, 0.03])
        pendencia.append(p)

    # Peso fixo de 5%
    peso_pendencias = [0.05] * num_amostras

    # Nota do indicador Pendências (1-5)
    nota_i8 = []
    for p in pendencia:
        if 'Sem pendências' in p:
            nota = 5
        elif 'Boa Vista' in p:
            nota = 4
        elif 'Pefin' in p or 'Dívida vencida' in p:
            nota = 3
        elif 'Protesto' in p or 'Ação Judicial' in p:
            nota = 2
        elif 'REFIN' in p:
            nota = 1
        else:
            nota = 1

        nota_i8.append(nota)

    # Nota ponderada
    nota_pi8 = [n * 0.05 for n in nota_i8]

    # Possui Garantia
    garantia = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins tendem a oferecer menos garantias
            g = np.random.choice(garantias, p=[0.6, 0.05, 0.1, 0.05, 0.1, 0.1])
        else:
            g = np.random.choice(garantias, p=[0.2, 0.25, 0.15, 0.15, 0.15, 0.1])
        garantia.append(g)

    # Peso fixo de 5%
    peso_garantia = [0.05] * num_amostras

    # Nota do indicador Garantia (1-5)
    nota_i9 = [5 if g != 'não' else 1 for g in garantia]

    # Nota ponderada
    nota_pi9 = [n * 0.05 for n in nota_i9]

    # Possui Seguros
    seguro = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins tendem a ter menos seguros
            s = np.random.choice(seguros, p=[0.6, 0.15, 0.15, 0.05, 0.05])
        else:
            s = np.random.choice(seguros, p=[0.1, 0.2, 0.2, 0.25, 0.25])
        seguro.append(s)

    # Peso fixo de 5%
    peso_seguros = [0.05] * num_amostras

    # Nota do indicador Seguros (1-5)
    nota_i10 = []
    for s in seguro:
        if s == 'não':
            nota = 1
        elif 'estoque' in s:
            nota = 3
        else:
            nota = 5
        nota_i10.append(nota)


    # Nota ponderada
    nota_pi10 = [n * 0.05 for n in nota_i10]

    # Serasa Score - Valores entre 0 e 1000
    serasa_score = []
    for i in range(num_amostras):
        if perfil_ruim[i]:
            # Empresas ruins têm score Serasa mais baixo
            base = np.random.beta(2, 5) * 1000  # Mais concentrado em valores baixos
        else:
            base = np.random.beta(5, 2) * 1000  # Mais concentrado em valores altos

        # Arredondar para 50s mais próximos para parecer mais realista
        score = round(base / 50) * 50
        serasa_score.append(int(min(1000, max(0, score))))

    # Peso fixo de 10%
    peso_serasa_score = [0.1] * num_amostras

    # Nota do indicador Serasa Score (1-5)
    nota_i11 = []
    for score in serasa_score:
        if 0 < score <= 300:
            nota = 1
        elif score <= 500:
            nota = 2
        elif score <= 700:
            nota = 3
        elif score <= 900:
            nota = 4
        elif score <= 1000:
            nota = 5
        else:
            nota = 1
        nota_i11.append(nota)

    # Nota ponderada
    nota_pi11 = [n * 0.1 for n in nota_i11]

    # Calcular Score GO ON (média ponderada das notas)
    score_go_on = []
    for i in range(num_amostras):
        score = (nota_pi1[i] + nota_pi2[i] + nota_pi3[i] + nota_pi4[i] +
                 nota_pi5[i] + nota_pi6[i] + nota_pi7[i] + nota_pi8[i] +
                 nota_pi9[i] + nota_pi10[i] + nota_pi11[i])
        score_go_on.append(round(score, 2))

    # Calcular Rating GO ON com base no Score
    rating_go_on = []
    for score in score_go_on:
        if score < 1.8:
            rating = 1
        elif score < 2.6:
            rating = 2
        elif score < 3.4:
            rating = 3
        elif score < 4.2:
            rating = 4
        else:
            rating = 5
        rating_go_on.append(rating)

    # Crédito Concedido (baseado no Rating)
    credito_concedido = []
    for rating in rating_go_on:
        if rating >= 3:
            credito_concedido.append("Sim")
        else:
            credito_concedido.append("Não")

    # Valor Limite de Crédito
    valor_limite_credito = []
    for i in range(num_amostras):
        if credito_concedido[i] == "Sim":
            # Fórmula: (EBITDA x Fator Multiplicador) - Dívidas
            limite = (ebitda[i] * multiplicador_setor[i]) - divida[i]
            # Garantir que o limite seja positivo
            limite = max(0, limite)
            valor_limite_credito.append(round(limite))
        else:
            valor_limite_credito.append(0)

    # Criar DataFrame com todos os dados
    df = pd.DataFrame({
        'Empresa': empresas,
        'EBITDA': [round(e) for e in ebitda],
        'Divida': [round(d) for d in divida],
        'Multiplicador do Setor': multiplicador_setor,
        'Margem EBITDA': [round(m, 2) for m in margem_ebitda],
        'Peso_margem_ebitda': peso_margem_ebitda,
        'nota_i1': nota_i1,
        'nota_pi1': nota_pi1,
        'Divida liquida/EBTIDA': [round(d, 2) for d in divida_liquida_ebitda],
        'peso_Divida liquida/EBTIDA': peso_divida_liquida_ebitda,
        'nota_i2': nota_i2,
        'nota_pi2': nota_pi2,
        'Indice de liquidez': [round(il, 2) for il in indice_liquidez],
        'peso_Indice de liquidez': peso_indice_liquidez,
        'nota_i3': nota_i3,
        'nota_pi3': nota_pi3,
        'Ciclo Financeiro': [round(cf) for cf in ciclo_financeiro],
        'peso_ciclo_financeiro': peso_ciclo_financeiro,
        'nota_i4': nota_i4,
        'nota_pi4': nota_pi4,
        'Conversão de Ebitda em FCO': [round(c, 2) for c in conversao_ebitda_fco],
        'peso_conversao_ebitda_em_fco': peso_conversao_ebitda_fco,
        'nota_i5': nota_i5,
        'nota_pi5': nota_pi5,
        'Tempo de atuação em anos': tempo_atuacao,
        'peso_tempo_atuacao': peso_tempo_atuacao,
        'nota_i6': nota_i6,
        'nota_pi6': nota_pi6,
        'Auditoria em Demonstrações Financeiras': auditoria,
        'peso_auditoria': peso_auditoria,
        'nota_i7': nota_i7,
        'nota_pi7': nota_pi7,
        'Pendencias Financeiras': pendencia,
        'peso_pendencias': peso_pendencias,
        'nota_i8': nota_i8,
        'nota_pi8': nota_pi8,
        'Possui Garantia': garantia,
        'peso_garantia': peso_garantia,
        'nota_i9': nota_i9,
        'nota_pi9': nota_pi9,
        'Possui Seguros': seguro,
        'peso_seguros': peso_seguros,
        'nota_i10': nota_i10,
        'nota_pi10': nota_pi10,
        'Serasa Score': serasa_score,
        'peso_serasa_score': peso_serasa_score,
        'nota_i11': nota_i11,
        'nota_pi11': nota_pi11,
        'SCORE GO ON': score_go_on,
        'RATING GO ON': rating_go_on,
        'Crédito Concedido?': credito_concedido,
        'Valor Limite de Crédito': valor_limite_credito
    })

    return df

# Gerar conjuntos de dados sintéticos
def gerar_e_salvar_dados(num_registros=50000, arquivo_saida="df_oficial.csv", taxa_reprovacao=0.40):
    # Gerar dados
    df_sintetico = gerar_dados_credito_empresarial(num_registros, taxa_reprovacao_alvo=taxa_reprovacao)

    # Salvar em CSV
    df_sintetico.to_csv(arquivo_saida, index=False)

    print(f"Dataset sintético gerado com {num_registros} registros e salvo em '{arquivo_saida}'")

    # Retornar algumas estatísticas básicas
    stats = {
        'total_registros': len(df_sintetico),
        'media_ebitda': df_sintetico['EBITDA'].mean(),
        'media_divida': df_sintetico['Divida'].mean(),
        'media_score': df_sintetico['SCORE GO ON'].mean(),
        'distribuicao_rating': df_sintetico['RATING GO ON'].value_counts().to_dict(),
        'taxa_aprovacao': (df_sintetico['Crédito Concedido?'] == 'Sim').mean() * 100,
        'taxa_reprovacao': (df_sintetico['Crédito Concedido?'] == 'Não').mean() * 100
    }

    print("\nEstatísticas do dataset gerado:")
    print(f"Total de registros: {stats['total_registros']}")
    print(f"Média Score GO ON: {stats['media_score']:.2f}")
    print(f"Distribuição de Rating: {stats['distribuicao_rating']}")
    print(f"Taxa de aprovação de crédito: {stats['taxa_aprovacao']:.1f}%")
    print(f"Taxa de reprovação de crédito: {stats['taxa_reprovacao']:.1f}%")

    return df_sintetico, stats

# Função para preparar dados para visualizações e análises
def preparar_analises(df):
    # Analisar a correlação entre as variáveis e o resultado de concessão
    resultado_credito = df['Crédito Concedido?'].map({'Sim': 1, 'Não': 0})

    # Selecionar colunas numéricas para análise de correlação
    cols_numericas = ['EBITDA', 'Divida', 'Margem EBITDA', 'Divida liquida/EBTIDA',
                      'Indice de liquidez', 'Ciclo Financeiro', 'Conversão de Ebitda em FCO',
                      'Tempo de atuação em anos', 'Serasa Score', 'SCORE GO ON', 'RATING GO ON']

    df_corr = df[cols_numericas].copy()
    df_corr['Concessão'] = resultado_credito

    # Calcular e exibir correlações
    correlacoes = df_corr.corr()['Concessão'].sort_values(ascending=False)
    print("\nCorrelações com a concessão de crédito:")
    print(correlacoes)

    return correlacoes

# Adicione código aqui para executar a geração de dados
if __name__ == "__main__":
    # Exemplo de uso com taxa de reprovação de aproximadamente 35%
    df, stats = gerar_e_salvar_dados(50000, taxa_reprovacao=0.40)

    # Analisar correlações
    correlacoes = preparar_analises(df)
````

````python
def validar_calculos(df):
    inconsistencias = {
        'score_go_on': [],
        'rating_go_on': [],
        'limite_credito': []
    }

    for idx, row in df.iterrows():
        # 1. Validar SCORE GO ON
        notas_ponderadas = [
            row['nota_pi1'], row['nota_pi2'], row['nota_pi3'], row['nota_pi4'],
            row['nota_pi5'], row['nota_pi6'], row['nota_pi7'], row['nota_pi8'],
            row['nota_pi9'], row['nota_pi10'], row['nota_pi11']
        ]
        score_calc = round(sum(notas_ponderadas), 2)
        if not np.isclose(score_calc, row['SCORE GO ON'], atol=0.01):
            inconsistencias['score_go_on'].append(idx)

        # 2. Validar RATING GO ON
        score = row['SCORE GO ON']
        if score < 1.8:
            rating_calc = 1
        elif score < 2.6:
            rating_calc = 2
        elif score < 3.4:
            rating_calc = 3
        elif score < 4.2:
            rating_calc = 4
        else:
            rating_calc = 5

        if rating_calc != row['RATING GO ON']:
            inconsistencias['rating_go_on'].append(idx)

        # 3. Validar Limite de Crédito
        if row['Crédito Concedido?'] == 'Sim':
            limite_calc = max(0, (row['EBITDA'] * row['Multiplicador do Setor']) - row['Divida'])
        else:
            limite_calc = 0

        if not np.isclose(limite_calc, row['Valor Limite de Crédito'], atol=1):
            inconsistencias['limite_credito'].append(idx)

    # Relatório
    print("\n📊 Validação dos Cálculos:")
    print(f"→ SCORE GO ON divergente em {len(inconsistencias['score_go_on'])} registros")
    print(f"→ RATING GO ON divergente em {len(inconsistencias['rating_go_on'])} registros")
    print(f"→ LIMITE DE CRÉDITO divergente em {len(inconsistencias['limite_credito'])} registros")

    return inconsistencias
````

````python
inconsistencias = validar_calculos(df)
````
