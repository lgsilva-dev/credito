```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
# Carregar dados
df = pd.read_csv('/content/drive/MyDrive/gocn/df_graph.csv')  # ajuste conforme necessário
```

```python
def plotar_distribuicao(coluna):
    fig, ax = plt.subplots(figsize=(10, 5))
    cores = ['#FF6B6B', '#4ECDC4']
    sns.kdeplot(data=df[df['Target'] == 0], x=coluna, fill=True, color=cores[0], label='Reprovado (0)', ax=ax)
    sns.kdeplot(data=df[df['Target'] == 1], x=coluna, fill=True, color=cores[1], label='Aprovado (1)', ax=ax)
    ax.set_title(f'Distribuição de {coluna} por Concessão de Crédito')
    ax.set_xlabel(coluna)
    ax.set_ylabel('Densidade')
    ax.grid(True)
    ax.legend()
    plt.tight_layout()
    return plt
```

```python
def plotar_indicador(df, indicador):
    # Mapeamento de indicadores para notas
    indicadores_para_notas = {
        'Margem EBITDA': 'nota_i1',
        'Divida liquida/EBTIDA': 'nota_i2',
        'Indice de liquidez': 'nota_i3',
        'Ciclo Financeiro': 'nota_i4',
        'Conversao Ebitda em FCO': 'nota_i5',
        'Tempo de atuação em anos': 'nota_i6',
        'Auditorias': 'nota_i7',
        'PEFIN': 'nota_i8',
        'Garantia': 'nota_i9',
        'Seguros': 'nota_i10',
        'Serasa Score': 'nota_i11'
    }
    if indicador not in indicadores_para_notas:
        raise ValueError(f"Indicador '{indicador}' não encontrado no mapeamento.")
    nota = indicadores_para_notas[indicador]
    plt.close('all')
    fig, ax = plt.subplots(figsize=(10, 5))
    sns.boxplot(data=df, x=nota, y=indicador, ax=ax)
    ax.set_title(f'Distribuição da {nota} por {indicador}')
    ax.set_xlabel('Nota')
    ax.set_ylabel(indicador)
    ax.grid(True)
    plt.tight_layout()
    return fig
```
